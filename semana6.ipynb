```python
import pandas as pd
import json
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Leer los datos del archivo JSON
data_raw = pd.read_json('Ventas_registro.json')
```

```python
# Desanidamos los productos para usar los valores anidados (nombre - cantidad - total)
data_flat = pd.json_normalize(data_raw.to_dict(orient='records'), record_path='productos', meta=['tipo_venta', 'fecha_venta', 'pago_total', 'cliente'])
```

```python
# Gr치fica de correlaci칩n (Diagrama de dispersi칩n)
sns.scatterplot(x='cantidad', y='total', data=data_flat)
plt.title("Relaci칩n entre cantidad y total")
plt.xlabel("Cantidad")
plt.ylabel("Total")
plt.show()
```

```python
corr = data_flat[['cantidad', 'total', 'pago_total']].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Matriz de Correlaci칩n")
plt.show()
```

```python
sns.pairplot(data_flat, vars=['cantidad', 'total', 'pago_total'], hue='tipo_venta')
plt.show()
```
